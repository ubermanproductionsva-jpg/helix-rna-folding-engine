```python
# DIAG_3 — DELETE INCOMPLETE STANFORD DATASET

import shutil
import os

TARGET_PATH = "/content/stanford-rna-3d-folding-2"

print("⚠️ Target for deletion:", TARGET_PATH)

# ----------------------------
# CHECK IF EXISTS
# ----------------------------
if os.path.exists(TARGET_PATH):
    print("📁 Folder exists. Calculating size...")

    def get_size(path):
        total = 0
        for dirpath, dirnames, filenames in os.walk(path):
            for f in filenames:
                fp = os.path.join(dirpath, f)
                if os.path.exists(fp):
                    total += os.path.getsize(fp)
        return total / (1024**3)

    size_gb = get_size(TARGET_PATH)
    print(f"📦 Folder size: {size_gb:.2f} GB")

    # ----------------------------
    # DELETE
    # ----------------------------
    print("\n🔥 Deleting folder...")
    shutil.rmtree(TARGET_PATH)

    print("✅ Folder deleted.")

else:
    print("❌ Folder does not exist.")

# ----------------------------
# VERIFY
# ----------------------------
print("\n🔍 Verifying...")

if not os.path.exists(TARGET_PATH):
    print("✅ Confirmed: folder removed")
else:
    print("❌ Deletion failed")
```

```python
# 2 — HELIX ALIGNED DATAFRAME BUILDER (FIXED SPLIT PATHS)

import pandas as pd

# ----------------------------
# PATHS
# ----------------------------
FEATURE_PATH = "/content/drive/MyDrive/3/20/26 ALL/HELIX_TRAINING_READY/BASELINE_FEATURE_TABLE_CLEAN_STRICT.parquet"

DELTA_PATH = "/content/drive/MyDrive/kaggle notebvook/3 15 26/3 16/3 17 2026/Kaggle all/Protenix data/colab FP protenix/HELIX_KAGGLE_DUMP/HELIX_KAGGLE_DUMP_EXPORT/3 20 26 ALL/BASELINE_DELTA_TABLE_CLEAN_STRICT.parquet"

print("📥 Loading datasets...")

df_feat = pd.read_parquet(FEATURE_PATH)
df_delta = pd.read_parquet(DELTA_PATH)

print("✅ Loaded:")
print("Features:", df_feat.shape)
print("Delta:", df_delta.shape)

# ----------------------------
# KEYS
# ----------------------------
KEYS = ["target_id", "residue_index", "chain", "copy"]

# ----------------------------
# SAFE MERGE
# ----------------------------
df_model = df_feat.merge(
    df_delta[KEYS + ["delta_dx", "delta_dy", "delta_dz"]],
    on=KEYS,
    how="inner",
    validate="one_to_one"
)

print("\n✅ Merge complete:", df_model.shape)

# ----------------------------
# FILTERS
# ----------------------------
before = len(df_model)

if "coord_is_finite" in df_model.columns:
    df_model = df_model[df_model["coord_is_finite"] == 1]

if "trainable_row" in df_model.columns:
    df_model = df_model[df_model["trainable_row"] == 1]

after = len(df_model)

print(f"\n🧹 Filtered rows: {before} → {after}")

# ----------------------------
# REMOVE FIRST RESIDUE
# ----------------------------
df_model = df_model[df_model["residue_index"] > 1]

# ----------------------------
# FINAL CHECK
# ----------------------------
print("\n🔬 Final validation:")
print("Null delta_dx:", df_model["delta_dx"].isna().sum())

print("\n🧾 Sample:")
print(df_model.head())

# ----------------------------
# SAVE
# ----------------------------
OUTPUT_PATH = "/content/df_model_aligned.parquet"
df_model.to_parquet(OUTPUT_PATH, index=False)

print(f"\n💾 Saved → {OUTPUT_PATH}")
print("\n📊 FINAL SHAPE:", df_model.shape)
```

```python
# 3 — HELIX GEOMETRY-CONSTRAINED RECONSTRUCTION

import numpy as np
import pandas as pd

print("🧬 Starting geometry-constrained reconstruction...")

df = pd.read_parquet("/content/df_model_aligned.parquet")

# ----------------------------
# PARAMETERS
# ----------------------------
MAX_TURN_ANGLE = np.pi / 4   # 45° limit
EPS = 1e-8

# ----------------------------
# HELPERS
# ----------------------------
def normalize(v):
    norm = np.linalg.norm(v) + EPS
    return v / norm

def angle_between(v1, v2):
    v1_u = normalize(v1)
    v2_u = normalize(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))

def limit_turn(prev_vec, new_vec):
    angle = angle_between(prev_vec, new_vec)
    if angle <= MAX_TURN_ANGLE:
        return new_vec

    # interpolate direction
    prev_dir = normalize(prev_vec)
    new_dir = normalize(new_vec)

    blended = prev_dir + new_dir
    return normalize(blended) * np.linalg.norm(new_vec)

# ----------------------------
# RECONSTRUCTION
# ----------------------------
reconstructed = []

for target_id, g in df.groupby("target_id"):

    g = g.sort_values("residue_index")

    # 🔥 CRITICAL FIX: anchor to TRUE FIRST POINT
    first = g.iloc[0]
    current = np.array([first.x_1, first.y_1, first.z_1])

    prev_vec = None

    for _, row in g.iterrows():

        delta = np.array([row.delta_dx, row.delta_dy, row.delta_dz])

        if prev_vec is not None:
            delta = limit_turn(prev_vec, delta)

        current = current + delta
        prev_vec = delta

        reconstructed.append({
            "target_id": row.target_id,
            "residue_index": row.residue_index,
            "x_pred": current[0],
            "y_pred": current[1],
            "z_pred": current[2],
            "x_true": row.x_1,
            "y_true": row.y_1,
            "z_true": row.z_1
        })

print("✅ Reconstruction complete")

df_recon = pd.DataFrame(reconstructed)

# ----------------------------
# ERROR METRICS
# ----------------------------
errors = np.sqrt(
    (df_recon.x_pred - df_recon.x_true)**2 +
    (df_recon.y_pred - df_recon.y_true)**2 +
    (df_recon.z_pred - df_recon.z_true)**2
)

print("\n📊 RECONSTRUCTION ERROR")
print("Mean:", errors.mean())
print("Median:", errors.median())
print("P90:", np.percentile(errors, 90))

# ----------------------------
# ENDPOINT DRIFT
# ----------------------------
endpoint = df_recon.groupby("target_id").tail(1)

endpoint_error = np.sqrt(
    (endpoint.x_pred - endpoint.x_true)**2 +
    (endpoint.y_pred - endpoint.y_true)**2 +
    (endpoint.z_pred - endpoint.z_true)**2
)

print("\n📊 ENDPOINT DRIFT")
print("Mean:", endpoint_error.mean())
print("Median:", endpoint_error.median())
print("P90:", np.percentile(endpoint_error, 90))
```

```python
# 3 — HELIX GEOMETRY v2 (TURN + MAGNITUDE + VELOCITY CONTROL)

import numpy as np
import pandas as pd

print("🧬 Starting HELIX geometry v2 reconstruction...")

df = pd.read_parquet("/content/df_model_aligned.parquet")

MAX_TURN = np.pi / 6        # tighter (30°)
MAX_STEP = 3.0              # limit step magnitude
SMOOTHING = 0.7             # velocity blending
EPS = 1e-8

def normalize(v):
    return v / (np.linalg.norm(v) + EPS)

def limit_step(v):
    mag = np.linalg.norm(v)
    if mag > MAX_STEP:
        return normalize(v) * MAX_STEP
    return v

def limit_turn(prev, new):
    if prev is None:
        return new

    prev_u = normalize(prev)
    new_u = normalize(new)

    dot = np.clip(np.dot(prev_u, new_u), -1, 1)
    angle = np.arccos(dot)

    if angle <= MAX_TURN:
        return new

    # blend directions
    blended = normalize(prev_u + new_u)
    return blended * np.linalg.norm(new)

reconstructed = []

for target_id, g in df.groupby("target_id"):
    g = g.sort_values("residue_index")

    first = g.iloc[0]
    current = np.array([first.x_1, first.y_1, first.z_1])

    prev_vec = None

    for _, row in g.iterrows():

        delta = np.array([row.delta_dx, row.delta_dy, row.delta_dz])

        # 1. limit magnitude
        delta = limit_step(delta)

        # 2. limit turn
        delta = limit_turn(prev_vec, delta)

        # 3. velocity smoothing (CRITICAL)
        if prev_vec is not None:
            delta = SMOOTHING * prev_vec + (1 - SMOOTHING) * delta

        current = current + delta
        prev_vec = delta

        reconstructed.append({
            "target_id": row.target_id,
            "residue_index": row.residue_index,
            "x_pred": current[0],
            "y_pred": current[1],
            "z_pred": current[2],
            "x_true": row.x_1,
            "y_true": row.y_1,
            "z_true": row.z_1
        })

print("✅ Reconstruction complete")

df_recon = pd.DataFrame(reconstructed)

# ----------------------------
# ERROR
# ----------------------------
errors = np.sqrt(
    (df_recon.x_pred - df_recon.x_true)**2 +
    (df_recon.y_pred - df_recon.y_true)**2 +
    (df_recon.z_pred - df_recon.z_true)**2
)

print("\n📊 RECONSTRUCTION ERROR")
print("Mean:", errors.mean())
print("Median:", errors.median())
print("P90:", np.percentile(errors, 90))

# ----------------------------
# ENDPOINT
# ----------------------------
endpoint = df_recon.groupby("target_id").tail(1)

endpoint_error = np.sqrt(
    (endpoint.x_pred - endpoint.x_true)**2 +
    (endpoint.y_pred - endpoint.y_true)**2 +
    (endpoint.z_pred - endpoint.z_true)**2
)

print("\n📊 ENDPOINT DRIFT")
print("Mean:", endpoint_error.mean())
print("Median:", endpoint_error.median())
print("P90:", np.percentile(endpoint_error, 90))
```

```python
# 7 — TRUTH RECONSTRUCTION (CORRECTED FOR YOUR DATA)

import numpy as np
import pandas as pd

print("🧬 Starting TRUTH reconstruction (corrected)...")

df = pd.read_parquet("/content/df_model_aligned.parquet")

reconstructed = []
debug_checks = []

for target_id, g in df.groupby("target_id"):
    g = g.sort_values("residue_index")

    # 🔒 TRUE START (x_1 IS TRUTH)
    first = g.iloc[0]
    current = np.array([first.x_1, first.y_1, first.z_1])

    reconstructed.append({
        "target_id": first.target_id,
        "residue_index": first.residue_index,
        "x_pred": current[0],
        "y_pred": current[1],
        "z_pred": current[2],
        "x_true": first.x_1,
        "y_true": first.y_1,
        "z_true": first.z_1
    })

    for i in range(1, len(g)):
        row = g.iloc[i]

        delta = np.array([
            row.delta_dx,
            row.delta_dy,
            row.delta_dz
        ])

        prev = current.copy()
        current = current + delta

        reconstructed.append({
            "target_id": row.target_id,
            "residue_index": row.residue_index,
            "x_pred": current[0],
            "y_pred": current[1],
            "z_pred": current[2],
            "x_true": row.x_1,
            "y_true": row.y_1,
            "z_true": row.z_1
        })

        # 🔍 delta sanity check
        prev_true = np.array([
            g.iloc[i-1].x_1,
            g.iloc[i-1].y_1,
            g.iloc[i-1].z_1
        ])
        curr_true = np.array([
            row.x_1,
            row.y_1,
            row.z_1
        ])

        true_delta = curr_true - prev_true
        delta_error = np.linalg.norm(true_delta - delta)

        debug_checks.append(delta_error)

print("✅ Reconstruction complete")

df_recon = pd.DataFrame(reconstructed)

# 📊 POSITION ERROR
pos_error = np.sqrt(
    (df_recon.x_pred - df_recon.x_true)**2 +
    (df_recon.y_pred - df_recon.y_true)**2 +
    (df_recon.z_pred - df_recon.z_true)**2
)

# 📊 DELTA CONSISTENCY
debug_checks = np.array(debug_checks)

print("\n📊 TRUTH RECON ERROR")
print("Mean:", pos_error.mean())
print("Median:", np.median(pos_error))
print("P90:", np.percentile(pos_error, 90))

print("\n📊 DELTA CONSISTENCY ERROR")
print("Mean:", debug_checks.mean())
print("Max:", debug_checks.max())

print("\n🧾 Sample:")
print(df_recon.head())
```
