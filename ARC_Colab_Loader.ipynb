```python
import zipfile
import os

zip_name = list(uploaded.keys())[0]

with zipfile.ZipFile(zip_name, 'r') as zip_ref:
    zip_ref.extractall('arc_data')

print('Extracted to arc_data/')
```

```python
import json

def load_json(path):
    with open(path) as f:
        return json.load(f)

task_maps = load_json('arc_data/canonical_task_maps.json')
memory = load_json('arc_data/memory_db.json')
grammar = load_json('arc_data/grammar_dataset.json')

print('Loaded datasets successfully')
```
